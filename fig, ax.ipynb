```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
```

```python
y=np.linspace(1, 10, 10)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.plot(x,y, ls='', marker='o')
ax.tick_params(labelsize=15, width=6, color='gray')
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(np.cumsum(np.random.randn(1,20)), ls='', marker='o', ms=11, color='purple')
ax.tick_params(labelsize=16, width=8)
ax.grid()
```
